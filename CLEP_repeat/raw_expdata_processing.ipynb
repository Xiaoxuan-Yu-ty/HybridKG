## Real original ADNI data
1. seperate metadata and expdata
2. explode expdata: rows with multiple gene symbols
3. drop NaN
4. Rename gene symbols with the same name by adding suffix

```python
df_adni_raw = pd.read_csv("../data/ADNI/ADNI_Gene_Expression_Profile.csv", header=None, dtype=str, index_col=0)
```

```python
df_adni_raw.head(10)
```

```python
subject_row = df_adni_raw.loc['SubjectID',3:]
subject_ids = subject_row.dropna().tolist()

# extract metadata
meta_rows = ["Phase", "Visit", "260/280", "260/230", "RIN", "Affy Plate","YearofCollection"]
metadata = (
    df_adni_raw[df_adni_raw.index.isin(meta_rows)]
    .iloc[:, 3:]
    .T
)

metadata.index = subject_ids
metadata.index.name = "SubjectID"

metadata
```

```python
subject_row = df_adni_raw.loc['SubjectID',3:]
subject_ids = subject_row.tolist()

# extract exp data
expdata = (
    df_adni_raw[df_adni_raw.index.str.contains("_at$")]
    .iloc[:, 1:]
)
expdata.columns = ['Symbol']+ subject_ids

# Explode the multiple gene symbols
# Split the string by ' || ', resulting in a list, then explode into rows
expdata['Symbol'] = expdata['Symbol'].str.split(' \|\| ')
expdata = expdata.explode('Symbol')
```

```python
exp_dropna = expdata.dropna()
exp_dropna
```

```python
# Rename duplicated gene symbols
# Create a cumulative count for each symbol to generate the suffix
# 'cumcount' generates: 0, 1, 2... for occurrences of the same string
counts = exp_dropna.groupby('Symbol').cumcount()
# Append suffix only if the count is > 0
exp_dropna['Symbol'] = np.where(
    counts > 0, 
    exp_dropna['Symbol'] + '.' + counts.astype(int).astype(str), 
    exp_dropna['Symbol']
)
exp_dropna
```

```python
exp = exp_dropna.set_index('Symbol')
symbol_meaning = exp.iloc[:,-1].to_dict()
exp = exp.iloc[:,:-1].T
exp
```

```python
exp.to_csv("../data/ADNI/adni_exp_realcleaned.csv")
```

```python
import json

# Open a file in write mode ('w') and dump the dictionary
with open('../data/ADNI/adni_symbol_meaning.json', 'w', encoding='utf-8') as f:
    json.dump(symbol_meaning, f, indent=4, ensure_ascii=False)
```

```python
from matplotlib_venn import venn3, venn2
import matplotlib.pyplot as plt
```

```python
def plot_venn(set1, set2, label1, label2, title=''):
    plt.figure(figsize=(10,6))
    venn = venn2((set1, set2), set_labels=(label1, label2))
    #plt.legend(handles=[red_patch, blue_patch, green_patch], loc='lower right', title="Legend")
    plt.title(title)
    plt.show()
```
