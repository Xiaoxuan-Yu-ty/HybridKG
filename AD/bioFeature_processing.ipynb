```python
def generate_Bioeatures(ad_path, health_path, output:str, merge=True, sep='\t'):
    df_ad = pd.read_csv(ad_path, sep=sep, index_col=0)
    df_health = pd.read_csv(health_path, sep=sep, index_col=0)
    df_ad = df_ad.loc[:, df_ad.nunique()>1]
    df_health = df_health.loc[:, df_health.nunique()>1]
    if merge:
        df = pd.concat([df_ad.reset_index(drop=True), df_health.reset_index(drop=True)], axis=1)
        df.index = df_ad.index
        df.to_csv(output)
        return df
    else:
        df_ad.to_csv(output)
        return df_ad
```

```python
import os
import torch
import torch.nn.functional as F
import numpy as np
import sys

# Add parent directory to path for imports
try:
    base_dir = os.path.dirname(os.path.abspath(__file__))
except NameError:
    base_dir = os.getcwd()
sys.path.append(os.path.dirname(base_dir))
from utils.graph_utils import (
    create_fuzzy_rules,
    load_graph,
    get_kg_features,
    prepare_pyg_data,
    create_fuzzy_rules, 
)

from fuzzy_models.fuzzy_models import get_fuzzy_model
from tqdm import trange
```

```python
G = load_graph("../datasets/two_classes/no_label_leakage/G_NormExpressionSubgraph_k10.pkl")
kg_features = get_kg_features("../datasets/bioFeatures/subgraph_features.csv")
print(kg_features.shape)
data=prepare_pyg_data(G=G,
                        kg_feature_path="../datasets/bioFeatures/subgraph_features.csv",
                        kg_features=True,
                        topological_features=False)

#print(data)

in_channels = data.x.size(1)
out_channels = int(data.y.max().item() + 1)
num_rules = kg_features.shape[1]
# prepare input for training
model = get_fuzzy_model(
    model_type='gcn',
    in_channels=in_channels,
    hidden_channels=128,
    out_channels=out_channels,
    num_rules=num_rules
)

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.005,
    weight_decay=5e-4
)

topo_features_np = data.topo_features.cpu().numpy()

centers, widths = create_fuzzy_rules(
topo_features_np,
num_rules=model.num_rules
)

with torch.no_grad():
    model.fuzzy_layer.centers.copy_(
        torch.tensor(centers, dtype=torch.float)
    )
    model.fuzzy_layer.log_sigmas.copy_(
        torch.log(torch.tensor(widths, dtype=torch.float))
    )
    
# train
for epoch in trange(200, desc="Training"):
        model.train()
        optimizer.zero_grad()

        out, _ = model(
            data.x,
            data.edge_index,
            topo_features=data.topo_features
        )

        loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
```
